```python
import random
from functools import reduce
```

```python
suits=('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks=('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values={'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
        
    def __str__(self):
        return self.rank+' of '+self.suit

```

```python
class player():
    def __init__(self,name):
        self.name=name #player's name
        self.cards_at_hand=[]
        self.cards_at_hand_1=[]
        self.cards_at_hand_2=[]
        self.bet=0
        self.bet_1=0
        self.bet_2=0
        self.total_money=1000
        
        
    def add_money(self,money):
        self.total_money+=money
        
    def place_bet(self,bet):
        self.bet+=bet
        self.total_money-=self.bet
        
    def double_bet(self):
        self.bet*=2
   
    def bet_split(self):
        self.bet_1=self.bet
        self.bet_2=self.bet
        self.total_money-=self.bet_2
        
    def add_cards(self,new_card):
        self.cards_at_hand.append(new_card)
        
    def add_cards_1(self,new_card):
        self.cards_at_hand_1.append(new_card)
        
    def add_cards_2(self,new_card):
        self.cards_at_hand_2.append(new_card)
        
    
    def total_value(self):
        return reduce(lambda a,b:a+b,map(lambda i:i.value, self.cards_at_hand))
    
    def split_value_1(self):
        return reduce(lambda a,b:a+b,map(lambda i:i.value, self.cards_at_hand_1))
    
    def split_value_2(self):
        return reduce(lambda a,b:a+b,map(lambda i:i.value, self.cards_at_hand_2))
    
    def split(self):
        self.add_cards_1(self.cards_at_hand[1])
        self.add_cards_2(self.cards_at_hand[0])
        
    def reset_cards(self):
        self.cards_at_hand.clear()
        
        
    def reset_cards_1(self):
        self.cards_at_hand_1.clear()
        
    def reset_cards(self):
        self.cards_at_hand_1.clear()
        
        
    
        
    
        
        
        

    
    
    def __str__(self):
        if self.name=='Dealer':
            return f'{self.name} has {len(self.cards_at_hand)} cards'
        
        else:
            return f'Player {self.name} has ${self.total_money} and {len(self.cards_at_hand)} cards'
```

```python
def bet_input():
    input_range=[50,100,250,500,1000,5000]
    m=0
    
    
    while m not in input_range:
        m=int(input("Place your bet (50,100,250,500,1000,5000)"))
        
        if m not in input_range:
                print("\nSorry, I didn't understand that one. Please enter in range of (50,100,250,500,1000,500)")
                correct_input=False
                
        elif m in input_range:
            return m
            correct_input=True
           
```

```python
class deck():
    deck_cards=[] #list of cards initialized
    
    def __init__(self):
        for suit in suits:  #for loop to iterate through player choice
            for rank in ranks:
                self.deck_cards.append(Card(suit,rank)) #each card with a suit and rank will be appended to the list
                
    def shuffle(self):
        random.shuffle(self.deck_cards) #cards shuffled
        
    def deal_one(self):
        return self.deck_cards.pop() #after shuffling one card needs to be dealt out.
```

```python
def no_of_players_ip():
    
    player_list=[]
    p=int(input("Enter number of players: "))
    
    if p==1:
        player_one=player('one')
        player_list.append(player_one)
    elif p==2:
        player_one=player('one')
        player_two=player('two')
        player_list.extend([player_one,player_two])
    elif p==3:
        player_one=player('one')
        player_two=player('two')
        player_three=player('three')
        player_list.extend([player_one,player_two,player_three])
        
    elif p==4:
        player_one=player('one')
        player_two=player('two')
        player_three=player('three')
        player_four=player('four')
        player_list.extend([player_one,player_two,player_three,player_four])
        
    return player_list
```

```python
def player_turn_input(p):
    choice=' '
    input_range=['Hit','Stand','Double Down','Surrender']
    correct_input=False

    while correct_input==False :
        
        #if the user enters valid input choice
        choice=input(f'Player {p.name} Hit,Stand,Split or Double\n')
        if choice in input_range:
            correct_input=True
            
        #if the user enters invalid choice    
        else:
            
            print("Sorry, invalid choice! Please choose from (Hindi or English)")
            correct_input=False
            
    return choice
```

```python
def check_win(dealer,player):
    d=dealer.total_value()
    p=player.total_value()
    if p>d and p<22:
        return True

    elif p<d and d<22:
        return False
    
    elif p<d and d>21:
        print(f"Dealer bust,player {player.name} wins")
        return False
    
    elif p==d:
        return f'Push'
```
